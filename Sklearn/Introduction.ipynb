SVM support vector classification. Estimator's constructor

```python
from sklearn import svm

clf = svm.SVC(gamma=0.001, C=100.)
```

```python
clf.fit(digits.data[:-1], digits.target[:-1])
```

```python
clf.predict(digits.data[-1:])
```
