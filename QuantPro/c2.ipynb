```python
import qiskit as q
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram
import matplotlib

%matplotlib inline
```

```python
statevec_simulator = q.Aer.get_backend("statevector_simulator")
qasm_sim = q.Aer.get_backend("qasm_simulator")
```

```python
def zrob_cos(obwod):
    cos = q.execute(obwod, backend=statevec_simulator)
    rezult = cos.result()
    statevec = rezult.get_statevector()
    
    n_qubits = obwod.num_qubits
    obwod.measure([i for i in range(n_qubits)], [i for i in range(n_qubits)])
    
    qasm_cos = q.execute(obwod, backend=qasm_sim, shots=1024).result()
    zlicz = qasm_cos.get_counts()
    
    return statevec, zlicz
```

```python
obw = q.QuantumCircuit(2, 2)
statevec, zlicz = zrob_cos(obw)
```

```python
plot_bloch_multivector(statevec)
```

```python
obw = q.QuantumCircuit(2, 2)
obw.h(0)
statevec, zlicz = zrob_cos(obw)
plot_bloch_multivector(statevec)
```

```python
obw = q.QuantumCircuit(2, 2)
obw.h(0)
obw.cx(0, 1)
statevec, zlicz = zrob_cos(obw)
plot_bloch_multivector(statevec)
```

```python
plot_histogram([zlicz], legend=['output'])
```
