```python
import numpy as np
import tensorflow as tf
import pandas as pd
from matplotlib import pyplot as plt
from tensorflow import keras
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['age', 'affordibility']],
                                                   df.bought_insurance, test_size=0.2,
                                                   random_state=25)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age'] / 100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] / 100
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation="sigmoid", 
                       kernel_initializer="ones", bias_initializer="zeros")
])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
his = model.fit(X_train_scaled, y_train, epochs=5000, verbose=1)
```

```python
print(his.history.keys())
```

```python
plt.plot(his.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
plt.plot(his.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
model.evaluate(X_test_scaled, y_test)
```

```python
def log_loss(y_true, y_preds):
    epsilon = 1e-15
    y_preds_new = [max(i, epsilon) for i in y_preds]
    y_preds_new = [min(i, 1-epsilon) for i in y_preds_new]    
    y_preds_new = np.array(y_preds_new)
    
    return -np.mean(y_true*np.log(y_preds_new)+(1-y_true)*np.log(1-y_preds_new))
```

```python
def gradient_descent(age, affordability, y_true, epochs, loss_threshold):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_preds = sigmoid(weighted_sum)
        loss = log_loss(y_true, y_preds)
        
        w1d = (1 / n) * np.dot(np.transpose(age), (y_preds - y_true))
        w2d = (1 / n) * np.dot(np.transpose(affordability), (y_preds - y_true))
        
        bias_d = np.mean(y_preds - y_true)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d
        
        print(f"Epoch: {i} | w1: {w1} | w2: {w2} | bias: {bias} | loss: {loss}")
        
        if loss <= loss_threshold:
            break
    return w1, w2, bias
```

```python
gradient_descent(X_train_scaled["age"], X_train_scaled["affordibility"], y_train,
                1000, 0.4631)
```
